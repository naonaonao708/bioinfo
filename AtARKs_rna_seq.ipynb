```python
import pandas as pd
import numpy as np 
from scipy import stats
```

```python
df = pd.read_excel('data/RNA-seq WT-TKO.xlsx')
```

```python
TKO_osmo_tpm = df[['Name','TKO_Osmo3-1_S13_R1_001 trimmed (GE) - TPM','TKO_Osmo3-2_S14_R1_001 trimmed (GE) - TPM','TKO_Osmo3-3_S15_R1_001 trimmed (GE) - TPM']]
TKO_osmo_tpm = TKO_osmo_tpm.set_index('Name')
```

```python
WT_osmo_tpm = df[['Name','WT_Osmo3-1_S28_R1_001 trimmed (GE) - TPM','WT_Osmo3-2_S29_R1_001 trimmed (GE) - TPM','WT_Osmo3-3_S30_R1_001 trimmed (GE) - TPM']]
WT_osmo_tpm = WT_osmo_tpm.set_index('Name')
```

```python
WT_osmo_tpm.mean(axis=1)
```

```python
TKO_cpm = df[['Name','TKO_Osmo3-1_S13_R1_001 trimmed (GE) - CPM','TKO_Osmo3-2_S14_R1_001 trimmed (GE) - CPM','TKO_Osmo3-3_S15_R1_001 trimmed (GE) - CPM']]
TKO_cpm = TKO_cpm.set_index('Name')
```

```python
WT_cpm = df[['Name','WT_Osmo3-1_S28_R1_001 trimmed (GE) - CPM','WT_Osmo3-2_S29_R1_001 trimmed (GE) - CPM','WT_Osmo3-3_S30_R1_001 trimmed (GE) - CPM']]
WT_cpm = WT_cpm.set_index('Name')
```

```python
osmo_df = pd.concat([TKO_cpm.mean(axis=1),WT_cpm.mean(axis=1)],axis=1)
osmo_df
```

```python
osmo_df.columns = ['TKO','WT']
osmo_df
```

```python
dt = osmo_df['WT'] - osmo_df['TKO']
dt.sort_values(ascending=False)
```
